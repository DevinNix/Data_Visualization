```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('company_sales_data.csv')

x = df["month_number"]
y1 = df["facecream"]
y2 = df["facewash"]
y3 = df["toothpaste"]
y4 = df["bathingsoap"]
y5 = df["shampoo"]
y6 = df["moisturizer"]
y7 = df["total_units"]
y8 = df["total_profit"]
```

Exercise 1: Read the total profits of all months and show it using a line plot

```python
plt.plot(x, y8)

plt.title("Company Profit Per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")

plt.xticks(x)

plt.show()
```

Exercise 2: Get the total profit of all months and create line plot with the following Style properties

-Line Style dotted 

-Line color red

-Show the legend in the lower right corner.

-X label name = Month Number

-Y label name = Units Sold

-Add a circle marker.

-Line width should be 3

```python
plt.plot(x,\
         y8, \
         ls = "--", \
         c = "r", \
         lw = '3',\
         marker = 'o', \
         mec = 'r', \
         mfc = 'k')

plt.title("Company Sales Data of Last Year")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend(["Profit Data of Last Year"], loc = "lower right")

plt.xticks(x)

plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).


```python
plt.plot(x, y1, \
         x, y2, \
         x, y3, \
         x, y4, \
         x, y5, \
         x, y6, \
         marker = 'o', \
         lw = '2')

plt.title("Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend(["Face cream Sales Data", \
            "Face wash Sales Data", \
            "Toothpaste Sales Data", \
            "Bathingsoap Sales Data", \
            "Shampoo Sales Data", \
            "Moisturizer Sales Data"], loc = "upper left")

plt.xticks(x)

plt.show()
```

Exercise 4: Read the toothpaste sales data for each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should .

```python
plt.scatter(x, y3)

plt.grid(ls = '--')

plt.title("Toothpaste Sales Over the Last Year")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend(["Tooth Paste Sales Data"], loc = "upper left")

plt.xticks(x)

plt.show()
```

Exercise 5: Read the face cream and face wash product sales data and show it using a bar chart
    
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
barWidth = 0.25

plt.bar(x, y1, width = 0.25)
plt.bar(x + 0.25, y2, width = -0.25)

plt.grid(ls = '--')

plt.title("Face Cream Vs. Face Wash Sales")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend(["Face Cream Sales Data", "Face Wash Sales Data"], loc = "upper left")

plt.xticks(x)

plt.show()
```

Exercise 6: Read the sales data for bathing soap for all months and show it using a bar chart.

```python
barWidth = 0.25

plt.bar(x, y4, width = 0.75)

plt.grid(ls = '--')

plt.title("Bathingsoap Sales")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")

plt.xticks(x)

plt.show()
```

Exercise 7: Calculate the total sales data for last year for each product and show it using a Pie chart

Note: In the Pie chart display Number of units sold per year for each product in percentage.

```python
yTotals = pd.Series([y1.sum(), y2.sum(), y3.sum(), y4.sum(), y5.sum(), y6.sum()])
yTotLabels = ["Face Cream", "Face Wash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]

plt.pie(yTotals, labels = yTotLabels, autopct='%1.1f%%')

plt.title("Total Units")

plt.show()
```

Exercise 9: Show the bathingsoap and face wash sales together using a subplot

```python
plt.subplot(2, 1, 1)
plt.plot(x, y4, c = 'k', marker = 'o')

plt.legend(["Bathingsoap"], loc = "lower right")

plt.subplot(2, 1, 2)
plt.plot(x, y2, c = 'r', marker = 'o')

plt.suptitle("Sales Comparison")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend(["Face Wash"], loc = "lower right")

plt.show()
```
