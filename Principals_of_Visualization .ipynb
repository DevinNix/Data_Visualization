# 10 Principals of Visualization

**Promt:**
Your boss asks you to create a visualization to compare face cream units and face wash units sold in the year 2021 using the provided data. Describe the 10 principals of data visualization and how they pertain to this problem.

**Principal 1:**
Make sure to use the most impactfull data from the set to get your point across.

***In This Case:***
For this promt I would need 3 major pieces of information face cream sales and face wash sales for the year and the year itself 2021.

**Prinicipal 2:**
Know the software you are working with.

***In This Case:***
For this situation we will be using python with 3 different libraies matplotlib, pandas and numpy. We will be loading a provided .csv file with more data than we need and using pandas to store the data that pertains to this prompt.

***Code:***

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

df = pd.read_csv('company_sales_data.csv')

x = df["month_number"]
y1 = df["facecream"]
y2 = df["facewash"]
```

**Principal 3:**
Use the correct geometry to descibe the information in an easy to understand and digestible format.

***In This Case:***
For the momment we will use a multi-line plot since we need to show data over time and need to compare 2 different pieces of data.

***Code:***

```python
plt.plot(x, y1, x, y2, marker = 'o')

plt.grid(ls = '--')

plt.title("Face Cream and Face Wash Units Sold in 2021")
plt.xlabel("Month by Number")
plt.ylabel("Units Sold")
plt.legend(["Face Cream Units", "Face Wash Units"])

plt.xticks(x)

plt.show()
```

**Principal 4:**
Make sure to use an effective color scheme.

***In This Case:***
For this graph we used blue and orange which are both eye catching colors and serve our purpose. We also used a grid to provide an easier viewing experience.

**Principal 5:**
Include any metric of uncertainty. 

***In This Case:***
Since this data is fairly cut and dry there is not much uncertainty to show. This could very well change if you knew that there was a system error when forming the numbers or another similar problem. In that case you would want to provide a +- value of some sort to your chart to represent that uncertainty.

**Principal 6:**
Use small multiples(subplots) if appropriate.

***In This Case:***
For this information the use of 2 charts would be better than 1 since we are comparing information.

***Code:***

```python
#Face Cream Plot
plt.subplot(2, 1, 1)
plt.plot(x, y1, marker = 'o')

plt.grid(ls = '--')
plt.legend(["Face Cream"], loc = "upper left")
plt.xticks(x)

#Face Wash Plot
plt.subplot(2, 1, 2)
plt.plot(x, y2, c = 'orange', marker = 'o')

plt.grid(ls = '--')
plt.legend(["Face Wash"], loc = "upper left")
plt.xticks(x)

#Labels
plt.suptitle("Face Cream and Face Wash Units Sold in 2021")
plt.xlabel("Month by Number")
plt.ylabel("Units Sold")

plt.show()
```

**Principal 7:**
Always distinguish models from figures.

***In This Case:***
Since this is a straight forward figure comparing 2 different pieces of data not much needs to be done here. If this was a scientific model or something similar there would need to be a detailed description of the model included with the visualizations.

**Principal 8:**
Include detailed standalone captions. These captions should convey your data as effectivly as your graph does.

***In This Case:***

Graph 1: Face cream sales started to rise in March unitll May reaching 3600 units sold in the month of May alone. After this point units sold dipped to their lowest in the month of October where only 1990 units were sold in the month. Units sold steadily increased through the rest of the year.

**Principal 9:**
Consider an infographic.

***In This Case:***

Get an independant figure review to see if your figure conveys the information you are trying to convey.
